# **Task 1:** Simulate one dice roll in Python
Use well-named variable(s), constants(s), string(s), the random module and good user interaction (input and output). 

```python
#this saves space, by only importing randinit, since that is all we need
from random import randint
roll = randint(1,6)
#variables look for one word, e.g
#    roll_dice, RollDice, rolldice
#    input can be empty input()
roll_dice=input("Do you want to roll? ")
#if statement you must specify for each part, e.g. or roll_dice == ("y")
if roll_dice == ("yes") or roll_dice == ("y"):
    print("Roll:", roll)
else:
    print("Goodbye.")
```

# Task 2:  Weather based IF Statement 
Write a program that tells the user what to wear (shorts, trousers, hat, coat, etc etc you choose) based on the temperature that they input. Use well-named variable(s), constants(s), string(s), type conversion, a complex IF statement, comparison and logical operators, and good input and output messages


```python
tem = float (input("What is the temperature? "))
if tem > 80:
    print("How are you alive")

elif tem > 50:
    print("stay in air conditioned room")
elif tem > 30:
    print("Too hot, wear light clothes.")
elif tem >15:
    print("wear")
elif tem > 0:
    print ("Wear a thicker coat, maybe a scarf.")
else:
    print ("Wear lots of layers, a wooly hat, a scarf, gloves, socks, and whatever you can get.")
```

# Task 2:  extras

1) Allow entry to a movie if the user is older than 18 or if they are accompanied by an adult

2) Allow the user to come to your party if they are a boy and they like pizza

3) Remind the user to take an umbrella if the temperature is higher than 17 and the chance of rain is higher than 80%


```python
#allow entry to a movie
age = int(input("How old are you? "))
if age < 18:
    pg = input("Do you have an adult with you? ")
    #remember to use == for equals, NOT =
    if (pg == "yes") or (pg == "y") :
        print("You may pass.")
    else:
        print("You may not pass.")
else:
    print("You may pass.")
```

```python
#allow user to come to party if they are a boy and they like pizza
gen = input("Are you a boy? ")
if (gen == "yes") or (gen == "y"):
    piz = input("Do you like pizza? ")
    if (piz == "yes") or (piz == "y"):
        print("You can come to my party!")
    else:
        print("You need to like pizza to come to my party!")
else:
    print("You can't come to my party if you aren't a boy.")
        
```

```python
# Remind the user to take an umbrella if the temperature is higher than 17 and the chance of rain is higher than 80%
tem = float(input("What is the temperature?"))
rai = float(input("What percentage chance of rain is there?"))
if (tem > 17) and (rai > 80):
    print("Don't forget to take your umbrella!")
else:
    print("Have a good day!")
```
